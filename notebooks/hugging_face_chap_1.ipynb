## 🤗 Huging Face Ch 1 - Transformaers

The Transformers <code>pipeline()</code> function is a basic object in the 🤗 Transformer's library.

The <code>pipeline()</code> function allows for the selection of a model which undertakes all pre and post processing.

It's an entry route into NLP using defined transformer models.

This notebook focusses on the various <code>pipeline()</code> functions by way of introduction to NLP technologies.

```python
from transformers import pipeline # Import the pipeline function from the transformers library
from pipeline_utils import pretty_print_pipeline_result
```

#### 🤗 Sentiment Analysis pipeline

The <code>pipeline("sentiment-analysis")</code> allows for basic sentiment analysis in English however it's generic and not fine tuned to a specific use case or process.  It provides largely binary or ternary responses of <code>Positive</code>, <code>Negative</code> or <code>Neutral</code>.

Example below.

```python
classifyer_type = pipeline("sentiment-analysis") # Create a sentiment analysis pipeline

result = classifyer_type(
    "We’ve had so many emails recently. I’m searching through them. " \
    "At the moment I can only find emails with the meals and activities that Tilly will be doing. " \
    "I have just emailed as I’m thinking we have not received the kit list yet and the health form. " \
    "The timings are normally on those forms") # Use the pipeline to analyse the sentiment of the input text

pretty_print_pipeline_result(result) # Print the result in a user-friendly format
```

#### 🤗 Zero Shot Classification pipeline

The <code>pipeline("zero-shot-classification")</code> function takes sentiment analysis a step further by allowing the definition of labels for classification, thereby reducing the reliance on model provided labels.

Example below.

```python
classifier = pipeline("zero-shot-classification") # Create a zero-shot classification pipeline

result = classifier(
    "Hey mate, hope you're well. My reading of it is...Presentation on Fri." \
    "10 mins presenting with 5 for Q&A. The presentation is formative." \
    "Feedback from the presentation can be used for any edits on the assignment which is due in March (also formative)." \
    "I think that's right!  I also think he's offered to read drafts of the assignment if folk wanted." \
    "From memory, the whole module is a pass / fail affair and not graded in the traditional sense.",

    candidate_labels = ["education", "assignment", "presentation", "marking"],
) 

# Use the pipeline to classify the input text into one of the candidate labels

pretty_print_pipeline_result(result) # Print the result in a user-friendly format
```

#### 🤗 Text Generation pipeline

The <code>pipeline("text-generation")</code> function takes a prompt and the selected model will generate remaining text.  This involves a degree of randomness and is similar to the predictive text function found in word processing and texting functions on laptops and phones.

Example below.

```python
text_generation = pipeline("text-generation") # Create a text generation pipeline

result = text_generation("I am your philosophy teacher and I am going to talk about existentialism. ", max_length=50) # Use the pipeline to generate text based on the input prompt

pretty_print_pipeline_result(result) # Print the generated text in a user-friendly format
```

#### 🤗 Models

The above have used a default model selecttion for processing the task presented.

The Hugging Face model hub allows for the selection of specific models if required - https://huggingface.co/models

These can be selected as an argument in defining the classifier

```python
text_generation = pipeline("text-generation", model="HuggingFaceTB/SmolLM2-360M") # Create a text generation pipeline and specify a different model to use for text generation

result = text_generation("I am your philosophy teacher and I am going to talk about existentialism. ", max_length=50) # Use the pipeline to generate text based on the input prompt

pretty_print_pipeline_result(result) # Print the generated text in a user-friendly format
```

#### 🤗 Mask Filling

The <code>fill-mask</code> pipeline attempts to fill in the blanks in a given set of words


```python
unmasker = pipeline("fill-mask") # Create a fill-mask pipeline

result = unmasker("I love you so much.  I want to give you a big sloppy <mask>.", top_k=2) # Use the pipeline to fill in the masked word in the input text

pretty_print_pipeline_result(result) # Print the filled-in text in a user-friendly format
```

#### 🤗 Named Entity Recognition

The <code>ner</code> pipeline attempts separate entites from a gioven body of text - entities can include persons, places, items etc,


```python
ner = pipeline("ner") # Create a named entity recognition pipeline

result = ner("My name is Patrick. I live in Wymondham.  I have a dog called Gary.  Gary likes playing with a ball.  I drive a grey car.") # Use the pipeline to identify named entities in the input text

pretty_print_pipeline_result(result) # Print the identified named entities in a user-friendly format
```

#### 🤗 Question Answering

The <code>question-answering</code> pipeline attempts to answer questions based on a given body of text - you supply the pipeline a question and a body of text and it will porvide a contextual answer from the two inputs.


```python
question_answerer = pipeline("question-answering") # Create a question answering pipeline

result = question_answerer(
    question = "What is the name of the dog?", 
    context = "My name is Patrick. I live in Wymondham.  I have a dog called Gary.  Gary likes playing with a ball.  I drive a grey car."
) # Use the pipeline to answer the question based on the provided context

pretty_print_pipeline_result(result) # Print the answer in a user-friendly format
```
